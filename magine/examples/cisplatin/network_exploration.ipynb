```python
import networkx as nx
%matplotlib inline
import sys
```

```python
from magine.networks.visualization.notebooks.view import render_graph, draw_cyjs
from magine.networks.subgraphs import Subgraph
```

```python
g = nx.read_gpickle('Data/cisplatin_based_network.p')
net_sub = Subgraph(g)
```

```python
print(len(g.nodes()))
print(len(g.edges()))
```

```python
from exp_data import exp_data
```

```python
nodes = set(g.nodes())
species_measure = exp_data.species.id_list
sig_species_measure = set(exp_data.species.sig.id_list)
overlap_measured = len(nodes.intersection(species_measure))
sig_overlap_measured = len(nodes.intersection(sig_species_measure))
print(overlap_measured)
print(float(overlap_measured)/len(nodes)*100.)
print(sig_overlap_measured)
print(float(sig_overlap_measured)/len(nodes)*100.)
```

```python
bax_n = net_sub.neighbors('BAX', upstream=True, downstream=True)
```

```python
render_graph(bax_n)
```

```python
draw_cyjs(bax_n)
```

```python
expand = net_sub.expand_neighbors(bax_n, nodes='CASP3', downstream=True)
draw_cyjs(expand)
```

```python
expand2 = net_sub.expand_neighbors(expand, nodes='PARP1', downstream=True)
draw_cyjs(expand2)
```

```python
BAX_to_AKT = net_sub.paths_between_pair('BAX', 'AKT1')
```

```python
render_graph(BAX_to_AKT)
```

```python
nx.write_gml(BAX_to_AKT, 'bax_to_akt.gml')
```
