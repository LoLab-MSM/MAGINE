```python
import networkx as nx
%matplotlib inline
from IPython.display import Image
from magine.networks.visualization import draw_cyjs, draw_igraph
from magine.networks.subgraphs import Subgraph
from exp_data import exp_data
```

```python
g = nx.read_gpickle('Data/cisplatin_based_network.p')
sg_generator = Subgraph(g)
measured = exp_data.species.sig.id_list
```

```python
start = sg_generator.neighbors(
    'ERCC4', 
    downstream=True, 
    include_only=measured
)
```

```python
draw_cyjs(start)
```

```python
expand_up = sg_generator.expand_neighbors(start, 'ATR', 
                                          downstream=True, 
                                          include_only=measured)
```

```python
draw_cyjs(expand_up)
```

```python
expand_up2 = sg_generator.expand_neighbors(
    expand_up, 'TP53', 
    downstream=True, 
    include_only=exp_data.species.id_list
)
```
